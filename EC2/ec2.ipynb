# Do que se trata o EC2?

O Amazon Elastic Compute Cloud (EC2) se trata de um serviço de IaaS que provê capacidade computacional sob demanda

Seus principais beneficios são a redução de custos de hardware e a agilidade para fazer deploys de aplicações

Podemos escalar sua capacidade de modo vertical alterando o tipo da instancia que está sendo usada em períodos de tempo específicos por exemplo

# Componentes e Características do EC2

## Instancias

Se tratam das nossas maquinas virtuais

## Amazon Machine Images (AMI)

São **templates** pré configurados para as instancias que englobam os componentes **necessários para usarmos elas como servidores**

**As AMI incluem os sistemas operacionais e softwares adicionais**

São específicas para cada:

+ Região;
+ SO;
+ Arquitetura do processador;
+ Tipo do root device;
+ Tipo da virtualização usada.

Launch permissions se tratam de um ponto a mais aos listados acima para se pensar quando pensado na seleção de um AMI

## Instance Types

Os tipos de instancias de um EC2 selecionam as configurações de hardware da nossa instancia

Temos tipos especializados para cada necessidade

As areas que uma instancia pode possuir são:

+ Configurações de CPU
+ Memoria
+ Armazenamento
+ Capacidade de Rede
+ Hardware Grafico

## Amazon EBS Volume

São **volumes de armazenamento persistentes** usando o Amazon Elastic Block Store (EBS)

## Instance Store Volume

São armazenamentos feitos de modo temporário

**Dados salvos aqui são perdidos quando a instancia é parada, hibernada ou terminada**

## Key Pairs

Se trata de uma forma segura de fazer login na instancia

A **Amazon** retém uma chave publica (**public key**) e nós, **usuários**, temos uma chave privada (**private key**)

## Security Groups

São Security Groups do VPC

Possuem a finalidade de firewall controlando:

+ Protocolos
+ Portas
+ Faixas de IPs que podem acessar a instancia
+ Faixas de IPs que a instancia pode acessar
