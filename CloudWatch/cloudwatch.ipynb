#  Amazon CloudWatch

##  O que ?
Um servio de monitoramento, observabilidade, governana e otimizao

Podemos com o CloudWatch atuar usando as seguintes features:
    
+ **Trace**
    + nos permite rastrear processos realizados

+ **Logs**
    + monitorar logs

+ **Metrics**
    + analisar metricas de servios

+ **Alarms**
    + setar alarmes em casos que necessitam de ateno

###  Reteno de Dados
+ **Armazena mtricas no intervalo de tempo de 15 meses**
+ **O armazenamento de logs  ilimitado em termos de intervalo de tempo mas pode ser customizado**

##  Para que usar?

###  Monitoramento de Servios
Podemos fazer uso dele para monitorar metricas de outros servios, como S3, StepFunctions ou Lambda, permitindo tirar concluses a respeito de **otimizaes** a serem feitas, por exemplo.

###  Centralizao de Logs
**Fazendo uso do CloudWatch Logs podemos centralizar o monitoramento via logs de sistemas, aplicaes e servios AWS**.

###  Performance e Sade Operacional
+ **Rastrear performance ao longo da arquitetura**
+ **Ter uma viso unificada da sade operacional**

##  Aplicaes
+ Detectar comportamentos anmalos
+ Setar alarmes
+ Visualizar logs e mtricas
+ Tomar aes de modo automatizado
+ Resoluo de problemas
+ Ter insights para otimizar um recurso



##  Como usar?
Por armazenar metricas de modo historico, podemos definir um baseline que  considerado comum e qualquer pico acima deste baseline  detectado e reportado.
