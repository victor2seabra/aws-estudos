# ☁️ Amazon CloudWatch

## 📋 O que é?
Um serviço de monitoramento, observabilidade, governança e otimização

Podemos com o CloudWatch atuar usando as seguintes features:
    
+ **Trace**
    + nos permite rastrear processos realizados

+ **Logs**
    + monitorar logs

+ **Metrics**
    + analisar metricas de serviços

+ **Alarms**
    + setar alarmes em casos que necessitam de atenção

### 📅 Retenção de Dados
+ **Armazena métricas no intervalo de tempo de 15 meses**
+ **O armazenamento de logs é ilimitado em termos de intervalo de tempo mas pode ser customizado**

## 🎯 Para que usar?

### 🔍 Monitoramento de Serviços
Podemos fazer uso dele para monitorar metricas de outros serviços, como S3, StepFunctions ou Lambda, permitindo tirar conclusões a respeito de **otimizações** a serem feitas, por exemplo.

### 📝 Centralização de Logs
**Fazendo uso do CloudWatch Logs podemos centralizar o monitoramento via logs de sistemas, aplicações e serviços AWS**.

### 📊 Performance e Saúde Operacional
+ **Rastrear performance ao longo da arquitetura**
+ **Ter uma visão unificada da saúde operacional**

## 🚀 Aplicações
+ Detectar comportamentos anômalos
+ Setar alarmes
+ Visualizar logs e métricas
+ Tomar ações de modo automatizado
+ Resolução de problemas
+ Ter insights para otimizar um recurso



## 🛠️ Como usar?
Por armazenar metricas de modo historico, podemos definir um baseline que é considerado comum e qualquer pico acima deste baseline é detectado e reportado.
