# 🛡️ AWS Lake Formation

## 📋 Definição
Se trata de um serviço que tem como principal funcionalidade o **controle de quem/o que pode acessar um determinado arquivo/tabela**.

## 🔐 Granularidade de Controle
Ele atua com **menor granularidade que o AWS IAM**.

## 🔄 Funcionalidades Adicionais
Além do controle de segurança e privacidade, o LakeFormation pode **orquestrar mudanças em buckets** do S3.

## 🎯 Serviços Suportados
Ele permite controlar acesso dos seguintes serviços:

+ **Glue**
+ **Athena**
+ **Redshift Spectrum**
+ **Quicksight**
+ **EMR**

## 🔗 Integração com Glue Catalog
O **Lake Formation faz uso do Glue Catalog** para, automaticamente, classificar dados em:

+ **data lakes**
+ **data sources**
+ **transforms** (?)
+ **targets**

## 🧩 Componentes
O Lake Formation é **composto de**:

+ **Catalogs**

+ **Databases**

+ **Tables**

+ **Views**

+ **Data Filters**

+ **Data Sharing**

+ **Crawlers**

+ **Data Permissions**

+ **LF-Tags and Permissions**

+ **Hybrid Access Mode**

+ **Data Locations**

+ **Administrative Roles and Tasks**

+ **Data Lake Locations**

+ **Application Integration Settings**

+ **IAM Identity Center Integration**
    + essa integração permite ao LakeFormation controlar acessos de metadados (CRUD de metadados) no Glue Data Catalog;
    + controle de acessos de dados subjacentes no S3;
    + para um dado user/role acessar um dado ele deve ter permissões via IAM E via LakeFormation.

+ **Data Catalog Settings**

+ **Blueprints**
