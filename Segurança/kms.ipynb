
# AWS Key Management Service (KMS) 

O KMS (Key Management Service) permite a criptografia e codificao de recursos de forma centralizada em uma ferramenta.

Com ele, podemos ter controle sobre **chaves criptografadas** para proteger nossos recursos.

Ele faz uso de **HSMs (hardware security modules)** para **proteger** a **confidencialidade** e **integridade** das **keys**.


## Tipos de Chaves 

Suas keys podem ser **simtricas ou assimtricas**.


## Operaes com Chaves de Criptografia no KMS 

Podemos realizar as seguintes operaes sobre uma chave de criptografia do KMS:

+ **Criar**
+ **Importar**
+ **Rotacionar**
+ **Desabilitar**
+ **Deletar** (irreversvel, isto , no conseguimos decodificar os objects codificados usando esta key)
+ **Auditar o uso de chaves criptografadas**


## Integrao e Auditoria 

O KMS integra com diversas ferramentas e servios AWS, mas  **nativamente integrado com o AWS CloudTrail**.

**O AWS CloudTrail possibilita ao KMS a capacidade de auditar quem usou uma chave, em quais recursos e quando isso foi feito.**

Ele possui integrao com **S3, EBS, RDS, EFS, entre outros**.


## Criptografia em Data Lakes na AWS 

Data lakes na AWS tm dois tipos de criptografia:

### Do Lado do Servidor (Server-Side Encryption) 

* Aqui temos criptografia em dados em repouso (**data-at-rest**) escritos no S3.
* Data-at-rest consistem em dados armazenados no S3, mas sem estarem sendo transmitidos ou sendo usados ativamente. **"So dados que s esto ali no S3."**
* A criptografia  feita a **nvel de objects**.
* Desse modo, para acessar um determinado objeto o seguinte procedimento  adotado:

    **Upload de um object no S3  Criptografia de escrita desse object  Armazenamento**

    **Request de leitura de um object no S3  Codificao do object  Leitura**

### Do Lado do Cliente (Client-Side Encryption) 

* Este modo de criptografia consiste em **criptografar** os dados **antes** da **escrita** dos mesmos no **S3**.

**O KMS atua coordenando ambos os tipos de criptografia.**


## Segurana Reforada para Data Lakes 

A segurana do data lake pode ainda ser reforada fazendo uso de servios como IAM, Amazon API Gateway e Amazon Cognito.


## Custo do KMS 

Seu **custo**  baseado na **quantidade e gerenciamento de AWS KMS keys** e na **quantidade** de **requisies** via API destas.

A AWS oferece o AWS KMS Free Tier que inclui um nmero de requests que podem ser feitos mensalmente sem custos.

## Componentes do AWS KMS 

### Customer Managed Keys (CMK) 
* Gerenciada pelo cliente da AWS.
*  a **camada mais externa** de criptografia de uma **key**.
* Pode ser **usada diretamente** ou pode ser usada para **criptografar outras keys**.

### Keys 
* So organizadas de **modo hierrquico**.
* Possuem **alias, key identifiers e key versions**.
* A hierarquia  a responsvel por permitir e gerenciar o controle das keys.

### AWS Owned Keys (AOK) 
*  gerenciada pelo servio da AWS.
* **Service operators** tm a capacidade de gerenciar seu **lifecycle e permisses**.
* **Oferece como vantagem a facilidade de compartilhar dados entre regies e contas sem precisar se preocupar com Key Permissions**.
* No somos cobrados pela sua existncia ou uso.
* **No podemos alterar sua policy, auditar suas atividades ou delet-las**.

### Key Administrators 
* Pessoas que podem gerenciar a Key.

### Key Policy 
*  uma **poltica baseada em recursos** que controla o **acesso** a uma dada Key **e quem pode gerenci-la**.

### Key Rotation 
* Se trata do processo de substituir uma Key existente por outra.
* Feito de **modo automtico anualmente** pela AWS.

### Envelope Encryption 
* **Uma prtica onde o dado  criptografado usando uma data key e depois esta key  criptografada usando AWS Key**.
* Prov uma camada a mais de segurana.

### Grant 
* Um instrumento da policy que **permite administradores** da AWS **usarem** as **KMS Keys em operaes criptografadas**.

### Alias 
* Um apelido dado  Key, tornando mais eficiente identificar e usar nas aplicaes.

### Key Material 
* Um material criptogrfico usado na criptografia e decodificao de dados.
* **Podem ser gerados pelo KMS ou ainda importados pelo user**.

### HSM (Hardware Security Modules) 
* Um **aparelho fsico** que **resguarda e gerencia Keys** para maior segurana de autenticao e **prov processamento criptogrfico**.
