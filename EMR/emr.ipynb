# 🚀 Amazon EMR

## 📋 Definição
Se trata de uma ferramenta cujo **framework é altamente distribuído** tornando ágil e eficiente o processamento de dados.

## 🎯 Uso
Usado em situações de criação de pipelines robustos de dados.

## 🔧 Modos de Uso
Podemos usar EMR de 3 diferentes modos:

+ **EC2**
+ **EKS**
+ **Studio**

## ⚙️ Tecnologia Base
Faz uso de **Apache Hadoop** para realizar a distribuição e processar os dados.

## 📈 Escalabilidade
Seu cluster (job flow) é elástico, isto é, **podemos escalar de acordo com a nossa necessidade**.

## 🛠️ Ferramentas Disponíveis
Disponibiliza as ferramentas do Apache Hadoop como por exemplo:

+ **Hive**
+ **Pig**
+ **Spark**
+ **Flink**
+ **Hudi**
+ **Hue**
+ **Livy**
+ **MXNet**
+ **Trino**
+ **TensorFlow**
+ **HBase**
+ **Zeppelin**

## 🏗️ Papéis na Infraestrutura
O EMR desempenha os seguintes **papeis na infraestrutura**:

+ **provisionamento**
+ **gerenciamento**
+ **manutenção**

## 🔗 Integração com S3
**É integrado diretamente com o S3** permitindo consumir seus dados das seguintes maneiras:

+ **input**
+ **output**
+ **armazenamento intermediário**

## 💰 Modelo de Custo
**O seu custo é baseado no tempo que o cluster está ativo**.

## 🖥️ Natureza dos Clusters
**Os clusters (job flow) são máquinas do tipo EC2**.

## 💾 Opções de Armazenamento
O armazenamento de dados do EMR pode ser feito de diferentes formas:

+ **HDFS (Hadoop File System)** onde seu armazenamento é mantido somente de acordo com as regras de lifecycle do cluster
+ **EMRFS (Elastic MapReduce File System) usando S3 como camada de dados** sendo sujeito às regras de lifecycle do bucket

## ✅ Vantagens da Separação
A **vantagem de separar o armazenamento do meio computacional do EMR** faz com que ele tenha escalabilidade independente quanto à armazenamento e recursos computacionais, dessa forma **podemos escalar um sem impactar o outro**.
