#  AWS Lambda

##  Definio
Se trata de um servio **serverless** da AWS.

##  Modelo de Execuo

###  Baseado em Eventos
Sua **execuo**  **baseada em eventos** (chamados por StepFunctions, alteraes em bucket do S3, alteraes no DynamoDB, chamadas HTTP pelo API Gateway, dentre outros).

###  Execuo por Horrios
Apesar de ser baseada em eventos, quando usada em conjunto do Amazon Event Bridge podemos programar execues baseadas em horrios.

##  Escalabilidade
Como se trata de um servio serverless ele **escala automaticamente** sob demanda executando mltiplas instncias do codigo **em paralelo conforme necessrio**.

##  Linguagens Suportadas
Ele aceita as seguintes linguagens de programao:

+ **Python**
+ **Node.js**
+ **Java**
+ **Go**
+ **Ruby**
+ **.NET**
+ **Rust**
+ **Custom Runtime API (qualquer linguagem)**

##  Limitaes de Tempo e Execuo

###  Timeout
Seu tempo de execuo  configurvel tendo um timeout at **15 minutos**.

###  Execues Simultneas
Possui como **limitao** a execuo simultnea de no **mximo 1000 execues por regio**.

##  Isolamento e Segurana
Cada funo Lambda  executada em um **ambiente isolado**, nos garantindo mais **segurana** e **independncia**.

##  Integraes
Este servio integra com diversos servios AWS, dando exemplo de alguns temos:

+ **S3**
+ **DynamoDB**
+ **SNS**
+ **SQS**
+ **CloudWatch**
+ **StepFunctions**
+ entre outros

##  Cold Start

###  Conceito
O **cold start**  um fator a ser considerado em aplicaes serverless. Basicamente ele consiste no **tempo e "esforo"** necessrio para **inicializar** o que desejamos, nesse caso um AWS Lambda.

###  Motivos do Cold Start
No caso do AWS Lambda ele pode acontecer pelos seguintes motivos:

+ **primeira invocao** da funo
+ funo **inativa** por **10-15 minutos**
+ **picos de trfego** que excedem containers disponveis
+ **deploy** de nova verso
+ **mudana de configurao**

###  Variaes do Cold Start
+ O **cold start** pode variar para cada **linguagem de programao**
+ O **cold start** depende tambm do **tamanho da funo**

##  Processo de Inicializao

###  Inicializao Completa (Cold Start)
Uma **inicializao completa** de uma Lambda  composta das seguintes etapas:

**Requisio  Criar Container  Baixar Cdigo  Inicializar Runtime  Executar Funo**

###  Reutilizao (Warm Start)
No caso de uma reutilizao (**warm start**) temos as seguintes etapas:

**Requisio  Executar Funo (container j pronto)**

Como no temos que realizar a inicializao numa **warm start** ele  **dependente** somente do **tempo de execuo** da funo.

##  Otimizaes e Boas Prticas

###  Importao de Bibliotecas
Como boas prticas e para otimizar o cdigo devemos **evitar fazer importao de bibliotecas dentro de funes**, e sim no escopo global.

###  Provisioned Concurrency
Podemos tambm **setar o nmero de containers prontos constantemente** usando `ProvisionedConcurrency`, desse modo o cold start  evitado ou ao menos minizado.

##  Modelo de Cobrana
Seu mtodo de **cobrana**  pautado em:

+ **tempo** de execuo (**(tempo + cold start) x memria alocada**)
+ **quantidade** de invocaes
