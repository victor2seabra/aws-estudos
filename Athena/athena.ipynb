# 🔍 Amazon Athena

## 📋 Definição
Se trata de um serviço **serverless** usado com o intuito de executar **queries diretamente** em objetos do **S3** usando SQL.

## ⚙️ Características Técnicas

### 💰 Custo e Escalabilidade
- O **custo** é baseado na **quantidade de dados extraídos por query**
- O Athena **escala automaticamente**, processando queries em **paralelo**

### 🔄 Processamento
Como é baseado em **Trino** o **processamento de queries** é feito de **modo distribuído**.

## ✅ Vantagens
Uma das grandes vantagens do Athena é permitir que **queries usando SQL** sejam executadas em **dados estruturados, não-estruturados e semi-estruturados**.

## 📁 Formatos de Arquivo Suportados
Os formatos de arquivo suportados pelo Athena são:

+ **Parquet**
+ **ORC**
+ **AVRO**
+ **Iceberg**
+ **Delta Lake**
+ **JSON**
+ **CSV**

### 🎯 Recomendações de Formato
São **recomendados** os formatos **Parquet e ORC** uma vez que são **colunares**, assim, suas queries são mais otimizadas e performáticas.

## 🚀 Funcionalidades

### 📊 Queries Pontuais
Podemos também realizar queries pontuais usando ANSI SQL sem necessitar de carregar os dados no Athena ou fazer uma agregação dos arquivos.

### 📈 Análise de Dados
**Podemos usar Pyspark e Spark SQL para analisar nossos dados dentro do Athena via notebooks**.

## 🔗 Integração
Athena **integra nativamente** com o **Glue Catalog** permitindo performar queries sobre os metadados catalogados referentes a cada tabela.
