#  Amazon Athena

##  Definio
Se trata de um servio **serverless** usado com o intuito de executar **queries diretamente** em objetos do **S3** usando SQL.

##  Caractersticas Tcnicas

###  Custo e Escalabilidade
- O **custo**  baseado na **quantidade de dados extrados por query**
- O Athena **escala automaticamente**, processando queries em **paralelo**

###  Processamento
Como  baseado em **Trino** o **processamento de queries**  feito de **modo distribudo**.

##  Vantagens
Uma das grandes vantagens do Athena  permitir que **queries usando SQL** sejam executadas em **dados estruturados, no-estruturados e semi-estruturados**.

##  Formatos de Arquivo Suportados
Os formatos de arquivo suportados pelo Athena so:

+ **Parquet**
+ **ORC**
+ **AVRO**
+ **Iceberg**
+ **Delta Lake**
+ **JSON**
+ **CSV**

###  Recomendaes de Formato
So **recomendados** os formatos **Parquet e ORC** uma vez que so **colunares**, assim, suas queries so mais otimizadas e performticas.

##  Funcionalidades

###  Queries Pontuais
Podemos tambm realizar queries pontuais usando ANSI SQL sem necessitar de carregar os dados no Athena ou fazer uma agregao dos arquivos.

###  Anlise de Dados
**Podemos usar Pyspark e Spark SQL para analisar nossos dados dentro do Athena via notebooks**.

##  Integrao
Athena **integra nativamente** com o **Glue Catalog** permitindo performar queries sobre os metadados catalogados referentes a cada tabela.
